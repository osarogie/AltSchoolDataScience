1. Read the is link https://en.wikipedia.org/wiki/List_of_Asian_countries_by_GDP_(PPP)_per_capita into a table/dataframe and show the first 10 rows of the List of Asian countries by GDP (PPP) per capita

Score = 10 points

```python
import pandas as pd

df = pd.read_html(
    "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_GDP_(PPP)_per_capita")[0]

df.head(10)
```

2. Change the column name "GDP per capita (Int$)" to "Gross Domestic Product"

Score = 10 points

```python
df.rename(columns={'GDP per capita (Int$)': 'Gross Domestic Product'}, inplace=True)
df.head(10)
```

3. Convert the data type of the values in "Gross Domestic Product" to Int

Score = 10 points

```python
df['Gross Domestic Product'] = pd.to_numeric(df['Gross Domestic Product'])
df.head(10)
```

4. Convert the column "Gross Domestic Product" to a list and give it an intuitive variable name

Score = 5

```python
asian_countries_gdps = list(df['Gross Domestic Product'])
asian_countries_gdps
```

5. What is the average of the elements in the list?

Score = 5

```python
sum(asian_countries_gdps) / len(asian_countries_gdps)
```
